```python
g = nx.DiGraph()
g.add_node(1)
g.add_node(2)
g.add_node(3)
g.add_edge(1,2)
g.add_edge(2,3)
g.add_edge(3,1)
nx.draw(g,with_labels=True)
plt.show()
```

```python
g = nx.DiGraph()
g.add_node(1)
g.add_nodes_from([2,3])
g.add_nodes_from(range(4,7))
g.add_edge(1,2)
g.add_edge(1,1)
g.add_edges_from([(2,3),(3,6),(4,6),(5,6)])
nx.draw(g,with_labels=True,node_size=850,node_color='red',width=5,edge_color='green')
plt.show()
```

```python
g=nx.DiGraph()
l=[[1,0,0,1,0,1,1],[0,1,1,1,1,1,0],[1,1,1,0,1,0,1]]
g.add_nodes_from([0,1,2])
for i in range(0,3):
    for j in range(0,3):
        if l[i][j] == 1:
            g.add_edge(i,j)
nx.draw(g,with_labels=True,node_size=850,node_color='red',width=5,edge_color='green')
plt.show()
```
